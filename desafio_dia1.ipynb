#5 - Mudando os dados das colunas
* data_renovacao
* data_emprestimo
* data_devolucao

Adicionando colunas horário:

* horario_renovacao
* horario_emprestimo
* horario_devolucao

```python
# Convertendo as colunas para o tipo datetime, verificando se o valor é diferente de "0"
df['data_renovacao'] = pd.to_datetime(df['data_renovacao'])
df['data_emprestimo'] = pd.to_datetime(df['data_emprestimo'])
df['data_devolucao'] = pd.to_datetime(df['data_devolucao'], errors='coerce')  # 'coerce' converte valores não parseáveis para NaT (Not a Time)

# Extraindo apenas o horário das colunas de data e criando novas colunas de horário
df.insert(df.columns.get_loc('data_renovacao') + 1, 'horario_renovacao', df['data_renovacao'].dt.strftime('%H:%M'))
df.insert(df.columns.get_loc('data_emprestimo') + 1, 'horario_emprestimo', df['data_emprestimo'].dt.strftime('%H:%M'))
df.insert(df.columns.get_loc('data_devolucao') + 1, 'horario_devolucao', df['data_devolucao'].dt.strftime('%H:%M'))

# Exibindo o DataFrame com as novas colunas de horário
print(df)
```

#6 - Análises do Dataset

##6.1 - A quantidade de empréstimos está aumentando ou diminuindo ao decorrer dos últimos anos?

```python
import matplotlib.pyplot as plt

# Convertendo a coluna de data de empréstimo para o tipo datetime, se necessário
df['data_emprestimo'] = pd.to_datetime(df['data_emprestimo'])

# Criando uma nova coluna para o ano de empréstimo
df['ano_emprestimo'] = df['data_emprestimo'].dt.year

# Contando o número de empréstimos por ano
emprestimos_por_ano = df.groupby('ano_emprestimo').size()

# Plotando o gráfico de linha
plt.figure(figsize=(10, 6))
emprestimos_por_ano.plot(marker='o', color='b')
plt.title('Quantidade de Empréstimos por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Empréstimos')
plt.grid(True)
plt.show()
```

##6.2 - Em quais bibliotecas do sistema estão a maior quantidade de empréstimos?

```python
# Contando o número de empréstimos em cada biblioteca
emprestimos_por_biblioteca = df['biblioteca'].value_counts()

# Encontrando a biblioteca com o maior número de empréstimos
biblioteca_mais_emprestimos = emprestimos_por_biblioteca.idxmax()
quantidade_mais_emprestimos = emprestimos_por_biblioteca.max()

print(f"A biblioteca com a maior quantidade de empréstimos é '{biblioteca_mais_emprestimos}' com {quantidade_mais_emprestimos} empréstimos.")
```

##6.3 - Qual o tema com mais empréstimos?

```python
# Contando o número de empréstimos em cada tema (colecao)
emprestimos_por_tema = df['colecao'].value_counts()

# Encontrando o tema com o maior número de empréstimos
tema_mais_emprestado = emprestimos_por_tema.idxmax()
quantidade_mais_emprestados = emprestimos_por_tema.max()

print(f"O tema mais emprestado é '{tema_mais_emprestado}' com {quantidade_mais_emprestados} empréstimos.")
```

##6.4 - Qual o tema menos empréstimos?

```python
# Encontrando o tema com o menor número de empréstimos
tema_menos_emprestado = emprestimos_por_tema.idxmin()
quantidade_menos_emprestados = emprestimos_por_tema.min()

print(f"O tema menos emprestado é '{tema_menos_emprestado}' com {quantidade_menos_emprestados} empréstimos.")
```

##6.5 - Análise de empréstimos ao longo do tempo
* Essa análise permite entender a tendência de empréstimos ao longo dos anos, meses ou semanas, identificando padrões sazonais ou tendências de crescimento ou declínio.

```python
# Análise de empréstimos ao longo do tempo
emprestimos_por_mes = df['data_emprestimo'].dt.to_period('M').value_counts().sort_index()
emprestimos_por_mes.plot(kind='bar', figsize=(25, 6))
plt.title('Número de Empréstimos por Mês')
plt.xlabel('Mês')
plt.ylabel('Número de Empréstimos')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

##6.6 - Análise de status de material emprestado
* Essa análise permite entender a distribuição do status dos materiais emprestados, identificando a proporção de materiais emprestados, em renovação, em atraso, etc.

```python
# Contando o número de empréstimos em cada status de material emprestado
status_material_emprestado = df['status_material'].value_counts()

# Plotando o gráfico de barras com a quantidade de empréstimos nas barras
plt.figure(figsize=(10, 6))
ax = status_material_emprestado.plot(kind='bar', color='skyblue')
plt.title('Distribuição do Status de Material Emprestado')
plt.xlabel('Status de Material')
plt.ylabel('Quantidade de Empréstimos')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Adicionando a quantidade de empréstimos nas barras
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```
