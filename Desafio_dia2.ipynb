#5 - Utilizando a coluna localizacao como base para adicionar coluna tema
* conforme especificação recebida por email

```python
# Definindo uma função para extrair o tema com base nos números na coluna "localizacao"
def extrair_tema(numero):
    if 0 <= numero <= 99:
        return "Generalidades. Ciencia e conhecimento"
    elif 100 <= numero <= 199:
        return "Filosofia e psicologia"
    elif 200 <= numero <= 299:
        return "Religiao"
    elif 300 <= numero <= 399:
        return "Ciencias sociais"
    elif 400 <= numero <= 499:
        return "Classe vaga. Provisoriamente nao ocupada"
    elif 500 <= numero <= 599:
        return "Matematica e ciencias naturais"
    elif 600 <= numero <= 699:
        return "Ciencias aplicadas"
    elif 700 <= numero <= 799:
        return "Belas artes"
    elif 800 <= numero <= 899:
        return "Linguagem, Lingua, Linguistica"
    elif 900 <= numero <= 999:
        return "Geografia, Biografia, Historia"
    else:
        return None  # Retorna None se o número não estiver dentro dos intervalos especificados


# Aplicando a função à coluna "localizacao" para criar uma nova coluna "tema"
df['tema'] = df['localizacao'].apply(extrair_tema)

# Exibindo as primeiras linhas dos dados para verificar se a nova coluna foi criada corretamente
print(df.head())
```
