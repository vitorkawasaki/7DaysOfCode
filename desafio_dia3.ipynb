#2 - A quantidade de empréstimos está aumentando ou diminuindo ao decorrer dos últimos anos?

```python
# Convertendo a coluna de data de empréstimo para o tipo datetime, se necessário
df['data_emprestimo'] = pd.to_datetime(df['data_emprestimo'])

# Criando uma nova coluna para o ano de empréstimo
df['ano_emprestimo'] = df['data_emprestimo'].dt.year

# Contando o número de empréstimos por ano
emprestimos_por_ano = df.groupby('ano_emprestimo').size()

# Plotando o gráfico de linha
plt.figure(figsize=(10, 6))
emprestimos_por_ano.plot(marker='o', color='b')
plt.title('Quantidade de Empréstimos por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Empréstimos')
plt.grid(True)
plt.show()
```

# Análise em relação à visualização gerada.

Mas em 2017 ate 2019 houve uma crescimento exponencial, mas em 2020 talvez por conta da pandemia, houve uma queda que o valor ficou próximo de 2015.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Suponha que você tenha os dados em duas listas ou arrays
data = pd.date_range(start='2022-01-01', end='2024-12-31', freq='D')
valor = np.random.randint(1, 100, len(data))

# Criando o DataFrame
df = pd.DataFrame({'data': data, 'valor': valor})

# Convertendo a coluna de data para o tipo datetime, se necessário
df['data'] = pd.to_datetime(df['data'])

# Agrupando por mês e calculando a média dos valores
df_agrupado = df.groupby(pd.Grouper(key='data', freq='M')).mean().reset_index()

# Visualizando os dados agrupados
print(df_agrupado)

# Plotando o gráfico de barras
plt.figure(figsize=(10, 6))

# Barras representando a média dos valores por mês
plt.bar(df_agrupado['data'], df_agrupado['valor'], color='skyblue', width=20)

# Adicionando rótulos e título
plt.xlabel('Mês')
plt.ylabel('Média dos Valores')
plt.title('Média dos Valores por Mês')
plt.grid(axis='y')

# Rotacionando os rótulos do eixo x para melhorar a visualização
plt.xticks(rotation=45)

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```
