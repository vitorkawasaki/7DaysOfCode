# Importando Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Carregando Dataset

```python
df = pd.read_csv('desafio_dia2.csv', sep=';', encoding='ISO-8859-1')
```

#Verificação das colunas

```python
df.columns
```

```python
# Filtrar os empréstimos realizados entre 2015 e 2020
df_filtrado = df[(df['ano_emprestimo'] >= 2015) & (df['ano_emprestimo'] <= 2020)]

# Calcular a quantidade de empréstimos por curso de graduação
emprestimos_por_curso = df_filtrado['tema'].value_counts()
```

```python
# Plotar o gráfico de barras com as linhas com os valores escritos
plt.figure(figsize=(10, 6))
emprestimos_por_curso.plot(kind='bar', color='skyblue')
plt.title('Quantidade de Empréstimos por Curso de Graduação (2015-2020)')
plt.xlabel('Curso de Graduação')
plt.ylabel('Quantidade de Empréstimos')

# Adicionar os valores acima das barras
for i, value in enumerate(emprestimos_por_curso):
    plt.text(i, value + 0.1, str(value), ha='center')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
