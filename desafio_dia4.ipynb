## Como se distribuem os empréstimos de exemplares pelos tipos de vínculo dos usuários?

```python
# Contagem dos empréstimos por tipo de vínculo de usuário
emprestimos_por_vinculo = df['tipo_vinculo_usuario'].value_counts()

# Plotando um gráfico de barras
emprestimos_por_vinculo.plot(kind='bar', color='skyblue')
plt.title('Distribuição de empréstimos por tipo de vínculo de usuário')
plt.xlabel('Tipo de Vínculo de Usuário')
plt.ylabel('Número de Empréstimos')
plt.xticks(rotation=45)
plt.show()
```

##Quais coleções são mais emprestadas?

```python
# Contagem dos empréstimos por coleção
emprestimos_por_colecao = df['colecao'].value_counts()

# Plotando um gráfico de barras das 10 coleções mais emprestadas
emprestimos_por_colecao.head(10).plot(kind='bar', color='lightgreen')
plt.title('As 10 coleções mais emprestadas')
plt.xlabel('Coleção')
plt.ylabel('Número de Empréstimos')
plt.xticks(rotation=90)
plt.show()
```

##Quais são as bibliotecas com mais ou menos quantidade de empréstimos?

```python
# Contagem dos empréstimos por biblioteca
emprestimos_por_biblioteca = df['biblioteca'].value_counts()

# Plotando um gráfico de barras das bibliotecas com mais empréstimos
emprestimos_por_biblioteca.plot(kind='bar', color='salmon')
plt.title('Empréstimos por Biblioteca')
plt.xlabel('Biblioteca')
plt.ylabel('Número de Empréstimos')
plt.xticks(rotation=90)
plt.show()
```

CDU - contagem e percentual por tema

```python
# Função para gerar tabela de frequência com percentuais
def tabela_frequencia(df, coluna):
    # Contagem dos valores
    contagem = df[coluna].value_counts()
    # Percentuais
    percentuais = (contagem / contagem.sum()) * 100
    # Criando a tabela
    tabela = pd.DataFrame({'Contagem': contagem, 'Percentual': percentuais})
    return tabela

# Tabela de frequência para os temas da CDU dos exemplares emprestados
tabela_temas = tabela_frequencia(df, 'tema')

# Mostrando a tabela
print("Tabela de Frequência e Percentual para os Temas da CDU:")
print(tabela_temas)
```
